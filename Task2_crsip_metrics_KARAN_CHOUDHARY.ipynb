```python
#splitting the independent and dependent variables.
X = data.iloc[:,:-1].values
y = data.iloc[:,8].values
```

```python
print("shape of X independent variabe ",X.shape,"\nshape of Y dependent variabe ",y.shape)
```

```python
print(data.groupby('Outcome').size())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle="True",random_state=0)
```

```python
print("X_train  size :",X_train.shape,"\nX_test  size",X_test.shape)
print("y_train size ",y_train.shape,"\ny_test size",y_test.shape)
```

```python
#applying various agorithm ofr better accuarcy
from sklearn.tree import DecisionTreeClassifier 

model1 = DecisionTreeClassifier ()
model1.fit(X_train, y_train)

actual =y_test
predicted =model1.predict(X_test)
results = confusion_matrix(actual, predicted)
print('Confusion Matrix')
print(results)
print('Accuracy Score :', accuracy_score(actual, predicted))
print('Report')
print(classification_report(actual,predicted))

```

```python
#applying various agorithm ofr better accuarcy
from sklearn.ensemble import RandomForestClassifier 

model2 = RandomForestClassifier()
model2.fit(X_train, y_train)


actual =y_test
predicted =model2.predict(X_test)
results = confusion_matrix(actual, predicted)
print('Confusion Matrix')
print(results)
print('Accuracy Score :', accuracy_score(actual, predicted))
print('Report')
print(classification_report(actual,predicted))

```

```python
from xgboost import XGBClassifier

model3 = XGBClassifier()
model3.fit(X_train, y_train)

actual =y_test
predicted =model3.predict(X_test)
results = confusion_matrix(actual, predicted)
print('Confusion Matrix')
print(results)
print('Accuracy Score :', accuracy_score(actual, predicted))
print('Report')
print(classification_report(actual,predicted))

```
